This notebook focuses on cleaning and processing data from the Sample Superstore dataset. The goal is to perform an in-depth exploratory data analysis (EDA) to uncover meaningful business insights and patterns within the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SampleSuperstore.csv", encoding= 'unicode_escape')
```

```python
print(df.head())
print(df.info())
```

Before starting the analysis, the dataset is previewed using .head() to view the first few records and .info() to inspect column data types and detect any missing or inconsistent values. This step helps determine the necessary cleaning operations before proceeding with visualization and analysis.

```python
df.dropna(inplace=True)
```

The bar chart below illustrates the total sales per product category, providing a clear overview of which categories generate the highest revenue. This helps identify the businessâ€™s top-performing product segments.

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Category")
print(category_sales)
category_sales.plot(kind='bar', title='Total Sales by Category', ylabel='Total Sales', xlabel='Category')
plt.show()
```

The following visualization highlights the top 10 sub-categories ranked by average profit. This allows us to see which specific product types contribute most effectively to overall profitability.

```python
sub_profit = df.groupby('Sub-Category')['Profit'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 Sub Categories by Average Profit")
print(sub_profit)
sub_profit.plot(kind='bar', title='Average Profit by Sub Category', ylabel='Average Profit', xlabel='Sub Category')
plt.show()
```

The pie chart below represents sales distribution by region, offering a quick snapshot of geographical performance and identifying key markets that drive sales.

```python
region_sales = df.groupby('Region')['Sales'].sum()
print("\nSales by Region:")
print(region_sales)
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region')
plt.ylabel('')
plt.show()
```

This scatter plot examines the correlation between discount rates and profit margins. By visualizing this relationship, we can determine whether higher discounts consistently lead to lower profits and identify any notable outliers.

```python
df.plot(kind='scatter', x='Discount', y='Profit', title='Discount vs Profit')
plt.show()
```

The analysis reveals several actionable insights regarding sales performance, profitability, and regional trends. These findings can support data-driven decision-making in areas such as pricing strategy, inventory optimization, and regional marketing focus.
